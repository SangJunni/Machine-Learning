```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
```

```python
train = pd.read_csv("train_data.csv")
test = pd.read_csv("test_data.csv")
submit = pd.read_csv("submission.csv")
```

```python
train
```

```python
train_x = train.drop(['price', 'model', 'ID'], axis =1)
train_y = train['price']
train_x
```

```python
one_hot = pd.get_dummies(train['model'])
train_x = pd.concat([train_x, one_hot], axis=1)
train_x
```

```python
sc=StandardScaler()
train_scaled = sc.fit_transform(train_x.loc[:,train_x.columns == 'mileage'])
train_x['mileage']= train_scaled
```

```python
model=KNeighborsRegressor(n_neighbors =3, weights="distance")
model.fit(train_x, train_y)
```

```python
test_x = test.drop(['model', 'ID'], axis =1)
one_hot_test = pd.get_dummies(test['model'])
test_x = pd.concat([test_x, one_hot_test], axis=1)
test_scaled = sc.fit_transform(test_x.loc[:,test_x.columns == 'mileage'])
test_x['mileage']= test_scaled
```

```python
pred_y =model.predict(test_x)
submit['price'] = pred_y
submit.to_csv('submission.csv', index = False)
```
