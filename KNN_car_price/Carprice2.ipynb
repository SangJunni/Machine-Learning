```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
```

```python
train = pd.read_csv("train_data.csv")
test = pd.read_csv("test_data.csv")
submit = pd.read_csv("submission.csv")
```

```python
train
```

```python
train_x = train.drop(['price', 'model', 'ID'], axis =1)
train_y = train['price']
train_x
```

```python
sc=StandardScaler()
train_scaled = sc.fit_transform(train_x.loc[:,train_x.columns == 'mileage'])
train_x['mileage']= train_scaled
```

```python
train_x
```

```python
train_scaled = sc.fit_transform(train_x.loc[:,train_x.columns == 'tax'])
train_x['tax']= train_scaled
```

```python
classle = LabelEncoder()
train_x['model']= classle.fit_transform(train['model'])
train_x
```

```python
model=KNeighborsRegressor(n_neighbors =8, weights="distance")
model.fit(train_x, train_y)
```

```python
test_x = test.drop(['model', 'ID'], axis =1)
test_scaled = sc.fit_transform(test_x.loc[:,test_x.columns == 'mileage'])
test_x['mileage']= test_scaled
test_scaled = sc.fit_transform(test_x.loc[:,test_x.columns == 'tax'])
test_x['tax']= test_scaled
classle = LabelEncoder()
test_x['model']= classle.fit_transform(test['model'])
```

```python
pred_y =model.predict(test_x)
submit['price'] = pred_y
submit.to_csv('submission_test2.csv', index = False)
```
